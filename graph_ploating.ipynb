```python
import matplotlib.pyplot as plt

# Data for BERT
bert_epochs = [1, 2, 3, 4]
bert_train_loss = [0.953900, 0.544600, 0.400700, 0.280800]
bert_val_loss = [0.625896, 0.523362, 0.516584, 0.562933]
bert_accuracy = [0.796875, 0.833333, 0.844697, 0.842330]

# Data for RoBERTa
roberta_epochs = [1, 2, 3, 4]
roberta_train_loss = [1.016600, 0.547000, 0.413200, 0.313700]
roberta_val_loss = [0.640750, 0.515009, 0.502740, 0.497674]
roberta_accuracy = [0.790720, 0.824811, 0.839489, 0.851326]

# Data for DistilBERT
distilbert_epochs = [1, 2, 3, 4]
distilbert_train_loss = [1.023400, 0.627200, 0.478200, 0.362900]
distilbert_val_loss = [0.700712, 0.562694, 0.544509, 0.557560]
distilbert_accuracy = [0.772254, 0.810133, 0.816761, 0.820549]

# Create subplots for Training Loss, Validation Loss, and Accuracy
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))

# Plot Training Loss
ax1.plot(bert_epochs, bert_train_loss, label='BERT', marker='o', linewidth=2)
ax1.plot(roberta_epochs, roberta_train_loss, label='RoBERTa', marker='o', linewidth=2)
ax1.plot(distilbert_epochs, distilbert_train_loss, label='DistilBERT', marker='o', linewidth=2)
ax1.set_title('Training Loss Across Epochs', fontsize=16)
ax1.set_xlabel('Epoch', fontsize=14)
ax1.set_ylabel('Training Loss', fontsize=14)
ax1.legend(fontsize=12)
ax1.grid(True)
ax1.tick_params(axis='both', labelsize=12)

# Plot Validation Loss
ax2.plot(bert_epochs, bert_val_loss, label='BERT', marker='o', linewidth=2)
ax2.plot(roberta_epochs, roberta_val_loss, label='RoBERTa', marker='o', linewidth=2)
ax2.plot(distilbert_epochs, distilbert_val_loss, label='DistilBERT', marker='o', linewidth=2)
ax2.set_title('Validation Loss Across Epochs', fontsize=16)
ax2.set_xlabel('Epoch', fontsize=14)
ax2.set_ylabel('Validation Loss', fontsize=14)
ax2.legend(fontsize=12)
ax2.grid(True)
ax2.tick_params(axis='both', labelsize=12)

# Plot Accuracy
ax3.plot(bert_epochs, bert_accuracy, label='BERT', marker='o', linewidth=2)
ax3.plot(roberta_epochs, roberta_accuracy, label='RoBERTa', marker='o', linewidth=2)
ax3.plot(distilbert_epochs, distilbert_accuracy, label='DistilBERT', marker='o', linewidth=2)
ax3.set_title('Accuracy Across Epochs', fontsize=16)
ax3.set_xlabel('Epoch', fontsize=14)
ax3.set_ylabel('Accuracy', fontsize=14)
ax3.legend(fontsize=12)
ax3.grid(True)
ax3.tick_params(axis='both', labelsize=12)

# Adjust layout and display
plt.tight_layout()
plt.savefig(f"images/models-training-validation-loss", dpi=600, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    'Model': ['BERT', 'RoBERTa', 'DistilBERT'],
    'Overall Accuracy': [0.849, 0.8684, 0.8205],
    'Weighted Precision': [0.8482, 0.8678, 0.8273],
    'Weighted Recall': [0.849, 0.8684, 0.8205],
    'Weighted F1-Score': [0.8461, 0.8669, 0.8197]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot settings
metrics = ['Overall Accuracy', 'Weighted Precision', 'Weighted Recall', 'Weighted F1-Score']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
bar_width = 0.2
x = np.arange(len(metrics))

# Create grouped bar plot
plt.figure(figsize=(12, 6))
for i, model in enumerate(df['Model']):
    plt.bar(x + i*bar_width, df.iloc[i, 1:], width=bar_width, label=model, color=colors[i])

plt.xlabel('Metrics', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.title('Model Performance Comparison', fontsize=14)
plt.xticks(x + bar_width, metrics, rotation=45, ha='right')
plt.ylim(0.8, 0.88)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Add value labels
for i, metric in enumerate(metrics):
    for j, model in enumerate(df['Model']):
        value = df.loc[j, metric]
        plt.text(x[i] + j*bar_width, value + 0.003, f'{value:.4f}', 
                 ha='center', va='bottom', fontsize=9)

plt.savefig(f"images/three_model_performance", dpi=600, bbox_inches='tight')
plt.show()
```
